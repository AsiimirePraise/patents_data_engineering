```python
#import libraries
import pandas as pd
import os
```

```python
#size of the files

files = [
    "g_patent.tsv",
    "g_patent_abstract.tsv", 
    "g_inventor_disambiguated.tsv",
    "g_location_disambiguated.tsv",
    "g_assignee_disambiguated.tsv"
]

for file in files:
    size = os.path.getsize(f"./data/{file}")
    print(f"{file}: {size / (1024**2):.1f} MB")
```

#### since the files are big to prevent crashing i will only load specific columns
But first i need to know the columns each tsv has 

```python
for file in files:
    print(f"\n=== {file} ===")
    df_peek = pd.read_csv(f"./data/{file}", sep="\t", nrows=0)
    print(f"Number of columns: {len(df_peek.columns)}")
    print(f"Columns: {df_peek.columns.tolist()}")
```

```python
# 1. Patents
df_patent = pd.read_csv("./data/g_patent.tsv", sep="\t",
                         usecols=["patent_id", "patent_title", "patent_date"],
                         dtype={"patent_id": str},
                         low_memory=False)
```

```python
# 2. Abstracts
df_abstract = pd.read_csv("./data/g_patent_abstract.tsv", sep="\t",
                           usecols=["patent_id", "patent_abstract"],
                           dtype={"patent_id": str},
                           low_memory=False)
```

```python
# 3. Inventors
df_inventor = pd.read_csv("./data/g_inventor_disambiguated.tsv", sep="\t",
                           usecols=["patent_id", "inventor_id", "disambig_inventor_name_first",
                                    "disambig_inventor_name_last", "location_id"],
                           dtype={"patent_id": str, "inventor_id": str, "location_id": str},
                           low_memory=False)
```

```python
# 4. Locations
df_location = pd.read_csv("./data/g_location_disambiguated.tsv", sep="\t",
                           usecols=["location_id", "disambig_country"],
                           dtype={"location_id": str},
                           low_memory=False)
```

```python

# 5. Assignees
df_assignee = pd.read_csv("./data/g_assignee_disambiguated.tsv", sep="\t",
                           usecols=["patent_id", "assignee_id", "disambig_assignee_organization"],
                           dtype={"patent_id": str, "assignee_id": str},
                           low_memory=False)
```

### save the extracted tsv

```python
# Create the folder
os.makedirs("./data/tsv_specific_columns", exist_ok=True)
```

```python
# 1. Patents
df_patent.to_csv("./data/tsv_specific_columns/patent.tsv", sep="\t", index=False)
print(f"patent.tsv saved → {df_patent.shape[0]:,} rows")
```

```python
# 2. Abstracts
df_abstract.to_csv("./data/tsv_specific_columns/abstract.tsv", sep="\t", index=False)
print(f"abstract.tsv saved → {df_abstract.shape[0]:,} rows")
```

```python
# 3. Inventors
df_inventor.to_csv("./data/tsv_specific_columns/inventor.tsv", sep="\t", index=False)
print(f"inventor.tsv saved → {df_inventor.shape[0]:,} rows")
```

```python
# 4. Locations
df_location.to_csv("./data/tsv_specific_columns/location.tsv", sep="\t", index=False)
print(f"location.tsv saved → {df_location.shape[0]:,} rows")
```

```python
# 5. Assignees
df_assignee.to_csv("./data/tsv_specific_columns/assignee.tsv", sep="\t", index=False)
print(f"assignee.tsv saved → {df_assignee.shape[0]:,} rows")
```

```python
print("\nAll specific column TSVs saved to ./data/tsv_specific_columns/")
```
